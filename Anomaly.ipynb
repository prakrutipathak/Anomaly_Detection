```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data =pd.read_csv("updated.csv", index_col=False)
```

```python
addata=data.where((pd.notnull(data)),'')
```

```python
addata.head()
```

```python
addata.shape
```

```python
addata.loc[addata['Labels'] == 'normal', 'Labels',] = 0
addata.loc[addata['Labels'] == 'abnormal', 'Labels',] = 1
```

```python
X=addata['Logs']
Y=addata['Labels']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase='True')

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

```python
# prediction on training data

prediction_on_training_data = model.predict(X_test_features)
accuracy_on_training_data = accuracy_score(Y_test, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
input = ["1118538965 2005.06.11 R30-M0-N9-C:J16-U01 2005-06-11-18.16.05.049256 R30-M0-N9-C:J16-U01 RAS KERNEL FATAL data TLB error interrupt"]
# convert text to feature vectors
input_data_features = feature_extraction.transform(input)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('abnormal')

else:
  print('normal')
```

```python
# Open the input CSV file
import csv
with open('test.csv', 'r') as input_file:
  # Create a CSV reader object
  reader = csv.reader(input_file)
  
  # Open the output CSV file
  with open('submission.csv', 'w') as output_file:
    # Create a CSV writer object
    writer = csv.writer(output_file)

    # Iterate over the rows in the input CSV
    for input in input_test_features:
  # Make a prediction for the current input
      prediction = model.predict(input)
      print("Prediction: ", prediction)

      if (prediction[0]==1):
        print('abnormal')
      else:
        print('normal')
      writer.writerow([prediction[0]])
```
